In Frequency Modulated Continuous Wave (FMCW) radar, the modulation pattern for changing frequencies is often implemented using either a **sawtooth** or **triangular** waveform. 

1. **Principle of Operation**:
   - FMCW radar continuously sends out a signal with a slowly changing frequency over time.
   - When this signal encounters a target, it reflects back toward the radar system.
   - The frequency of the reflected signal changes based on the target's distance and velocity.
   - By analyzing the time delay between transmission and reception, we can determine the target's range.

2. **Characteristics of FMCW Radar**:
   - FMCW radar systems often use a **sawtooth** pattern for frequency modulation. This pattern is crucial for accurate range and speed measurements, especially in applications like automotive radar¹.
   - The Doppler effect causes additional frequency shifting of the echo signal due to the target's motion. The relationship for the received frequency shift (\(f_{Rx}\)) due to Doppler effect is given by:
     $ f_{Rx} = f_{Tx} \left(1 + \frac{v}{c}\right) \left(1 - \frac{v}{c}\right) $
     where:
     - $f_{Rx}$ is the received frequency.
     - $f_{Tx}$ is the transmitted frequency.
     - \(v\) is the target's velocity.
     - \(c\) is the speed of light².

3. **Range Resolution**:
   - The range resolution of FMCW radar depends on the bandwidth of the transmitted FMCW signal.
   - The smaller the frequency intervals covered by the signal, the better the resolution.
   - The range resolution ($\Delta R$) is inversely proportional to the bandwidth (\(B_w\)) of the transmitted signal:
     $ \Delta R \propto \frac{1}{B_w} $
     Larger bandwidths lead to finer resolution⁵.

## FMCW Modulated Signal (Sawtooth)

The FMCW modulated signal (sawtooth) can be expressed as:

$ s_{\text{FMCW}}(t) = A_{\text{FMCW}} e^{j(2\pi (f_c + k_{\text{tr}}t) t + \phi_{\text{FMCW}})} $

Where:
- $f_c$ is the carrier frequency.
- $A_{\text{FMCW}}$ is the amplitude of the FMCW signal.
- $k_{\text{tr}}$ represents the frequency shift due to time delay.
- $\phi_{\text{FMCW}}$ is the phase of the FMCW signal.



## FMCW Radar and OFDM Combination

### 1. **OFDM Basics**
- **Orthogonal Frequency Division Multiplexing (OFDM)** is a modulation technique used in wireless communication systems.
- It divides the available bandwidth into multiple narrow subcarriers, each carrying data independently.
- The OFDM signal is generated by summing these subcarriers.

### 2. **Mathematical Description of OFDM**
- Consider an OFDM system with \(N\) subcarriers. Each subcarrier has a frequency \(f_k\) and a complex envelope \(s_k(t)\).
- The OFDM signal \(s(t)\) can be expressed as:
  $ s(t) = \sum_{k=0}^{N-1} s_k(t) $
- The complex envelope of each subcarrier \(s_k(t)\) can be represented as:
  $ s_k(t) = A_k e^{j(2\pi f_k t + \phi_k)} $
  where:
  - $A_k$ is the amplitude of the subcarrier.
  - $f_k$ is the frequency of the subcarrier.
  - $\phi_k$ is the phase of the subcarrier.

### 3. **FMCW Modulation**
- FMCW radar modulates the transmitted signal's frequency over time.
- Let's denote the frequency shift due to time delay as $k_{\text{tr}}$.
- The modulated FMCW signal can be expressed as:
  $ s_{\text{FMCW}}(t) = A_{\text{FMCW}} e^{j(2\pi (f_c + k_{\text{tr}}t) t + \phi_{\text{FMCW}})} $
  where:
  - $f_c$ is the carrier frequency.
  - $A_{\text{FMCW}}$ is the amplitude of the FMCW signal.
  - $\phi_{\text{FMCW}}$ is the phase of the FMCW signal.

### 4. **Combined Signal**
- To combine OFDM and FMCW, we multiply the OFDM signal by the FMCW signal:
  $ s_{\text{combined}}(t) = s(t) \cdot s_{\text{FMCW}}(t) $
- The resulting signal will have both OFDM subcarriers and the FMCW modulation.

### 5. **Important Considerations**
- Designing such a combined waveform requires careful consideration of spectral properties, orthogonality, and synchronization.
- The actual implementation involves more complex details, including guard intervals, synchronization, and channel estimation.




## Combined Signal of OFDM and FMCW

The combined signal results from multiplying the OFDM signal by the FMCW signal. Let's denote the OFDM signal as $s(t)$ and the FMCW signal as $s_{\text{FMCW}}(t)$. The combined signal $s_{\text{combined}}(t)$ can be expressed as:

$ s_{\text{combined}}(t) = s(t) \cdot s_{\text{FMCW}}(t) $

Where:
- $s(t)$ represents the OFDM signal.
- $s_{\text{FMCW}}(t)$ represents the FMCW signal.

This combined waveform integrates both OFDM subcarriers and the FMCW modulation. The design and implementation of such a waveform involve considerations related to spectral properties, orthogonality, and synchronization.

1. **Signal Reception**:
   - The receiver captures the combined signal, which includes both the OFDM subcarriers and the FMCW modulation.
   - The received signal may be affected by channel impairments, noise, and interference.

2. **Channel Estimation**:
   - The receiver estimates the channel frequency response to compensate for channel effects.
   - Techniques like pilot symbols or training sequences embedded in the OFDM signal help estimate the channel response.
   - Channel estimation allows the receiver to equalize the received signal.

3. **OFDM Demodulation**:
   - The receiver performs demodulation on the OFDM subcarriers.
   - This involves:
     - Removing the cyclic prefix (if present).
     - Applying the inverse fast Fourier transform (IFFT) to convert the frequency-domain OFDM symbols back to time-domain samples.
     - Equalizing the subcarriers using the estimated channel response.

4. **FMCW Demodulation**:
   - To detect the Doppler frequency (velocity) and range (distance), the receiver focuses on the FMCW component.
   - The FMCW modulation results in a beat frequency in the received signal.
   - The receiver extracts this beat frequency by mixing the received signal with a reference signal (generated locally).
   - The beat frequency corresponds to the Doppler shift due to the target's motion.

5. **Doppler Frequency Extraction**:
   - The beat frequency is proportional to the Doppler frequency:
     $ f_{\text{Doppler}} = \frac{{2 v f_c}}{{c}} $
     where:
     - \(v\) is the target velocity.
     - \(f_c\) is the carrier frequency.
     - \(c\) is the speed of light.

6. **Range Determination**:
   - The delay time (\(\Delta t\)) between the transmitted and received signals corresponds to the round-trip travel time.
   - The range to the target (\(R\)) can be calculated as:
     $ R = \frac{{c \Delta t}}{2} $

7. **Combined Processing**:
   - The receiver combines the information from both OFDM and FMCW components to obtain accurate range and velocity measurements.
   - Iterative processing may be used to refine the estimates.

8. **Bit Error Rate (BER) Calculation**:
   - The receiver evaluates the BER of the decoded OFDM symbols to assess communication performance.
